## Encoding

---

- `term` Encode target value (`36 months` = 36 & `60 months` = 60)
- `verification_status` Encode target value (`Not Verified` = 0 & `Verified` = 1)
- `initial_list_status` Encode target value (`f` = 0 & `w` = 1)
- `home_ownership` Encode target value (`OTHER` = 0, `OWN` = 1, & `RENT` = 2, `MORTGAGE` = 3)
- `grade` Encode target value (`A` = 1, `B` = 2, `C` = 3, `D` = 4, `E` = 5, `F` = 6, & `G` = 7)
- `emp_length` Encode target value (`0-5 years` = 0, `6-9 years` = 1, & `10+ years` = 2)
---

```python
from sklearn import preprocessing
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def column_encoding(data):

  label_encoder = preprocessing.LabelEncoder()
  
  data['purpose'] = label_encoder.fit_transform(data['purpose'])

  term = {'term' : {' 36 months' : 36, ' 60 months' : 60}}
  verification_status = {'verification_status' : {"Not Verified" : 0, "Verified" : 1}}
  initial_list_status = {'initial_list_status' : {"f" : 0, "w" : 1}}
  home_ownership = {'home_ownership' : {'OTHER' : 0, 'OWN' : 1, 'RENT' : 2, 'MORTGAGE' : 3}}
  grade = {'grade' : {"A" : 1, "B" : 2, "C" : 3, "D" : 4, "E" : 5, "F" : 6, "G" : 7}}
  emp_length = {'emp_length' : {"0-5 years" : 0, "6-9 years" : 1, "10+ years" : 2}}
  
  columns = [term, verification_status, initial_list_status, home_ownership, grade, emp_length]
  
  for column in columns:
    data.replace(column, inplace=True)

  return data
```

```python
train = column_encoding(train)

test = column_encoding(test)
```

```python
#Merging Xtrain
Xtrain = pd.concat([scaled_Xtrain,categorical_var_train],axis=1)
Xtrain = Xtrain.drop(['index','index'],axis=1)

#Merging Xtest
Xtest = pd.concat([scaled_Xtest,categorical_var_test],axis=1)
Xtest = Xtest.drop(['index','index'],axis=1)
```
